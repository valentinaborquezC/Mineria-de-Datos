# Caso Mineria de Datos

Australia es uno de los países más grandes del mundo y el continente más seco y plano del planeta. Aproximadamente la mitad occidental de Australia es una extensa meseta árida. El clima predominante es el desértico y semiárido (se estima que el 40% del territorio lo forman dunas de arena), aunque el norte del país cuenta con un clima tropical. Las tierras más fértiles y el clima templado con cuatro estaciones se concentran en el sureste y suroeste. La principal cordillera del país, la Gran Cordillera Divisoria (The Great Dividing Range), se extiende a lo largo de más de 3.500 km entre los estados de Queensland y Victoria. El pico más alto es el Monte Kosciuszko con 2.228 metros. Los ríos más caudalosos nacen en la Gran Cordillera Divisoria y, en su gran mayoría, fluyen hacia el este y desembocan en el océano Pacífico. Por su longitud y caudal destacan dos ríos: el Murray y el Darling.

Australia se divide en seis estados (New South Wales, Victoria, Queensland, South Australia, Western Australia y Tasmania) y dos territorios continentales (Northern Territory y Australian Capital Territory (ACT)). La ciudad de Sídney, con sus 5,73 millones de habitantes, es la más poblada de Australia y es la capital del estado de Nueva Gales del Sur. Es a su vez un importante centro industrial y puerto comercial de primera magnitud. La ciudad es muy extensa, con un radio aproximado de 40 km, incluyendo los suburbios. Otras ciudades importantes son Melbourne (5,19 millones), Brisbane (2,56 millones), Perth (2,38 millones) y Adelaide (1,41 millones). Casi el 90% de la población se concentra en las zonas urbanas. De sus más de 25 millones de habitantes, un 65,5% se encuentra en edad de trabajar (de 15 a 64 años), mientras que aquellos con más de 65 años y con menos de 15 años suponen el 15,7% y el 18,8% de la población respectivamente.

La superficie del país es de aproximadamente 7.741.220 km², quince veces superior a España, representando el 5% del total de la superficie terrestre emergida. Es el sexto país del mundo en extensión. La distancia máxima del continente australiano de este a oeste es de 4.100 km y de norte a sur, de 3.200 km. La longitud costera es de 36.735 km. Su vasta extensión inevitablemente condiciona el tráfico de personas, mercancías y servicios. Australia es, además, el más llano de los continentes, con una altitud media de menos de 300 metros, siendo The Great Dividing Range la única gran cordillera de importancia. Casi dos tercios del territorio carecen de corrientes de agua hacia el mar. La principal cuenca fluvial es la del Murray-Darling con 1.061.469 km², que drena las vertientes sur occidentales de la Gran Cordillera.

El país cuenta con una gran diversidad climática que abarca desde el clima tropical en el norte, que representa el 39% del territorio, al clima templado-continental

```python
# Importar las dependencias Necesarias

import pandas as pd   # para la manipulación de datos en forma de DataFrames
import numpy as np    # para operaciones numéricas y matemáticas
import matplotlib.pyplot as plt  # para la visualización de datos en gráficos
import seaborn as sns


# Cargar el conjunto de datos
data_frame = pd.read_csv('weatherAUS.csv')
```

```python
## Este bucle itera a través de todas las columnas del DataFrame
for feature in data_frame.columns:
    ## Imprime el nombre de la columna y el número de valores nulos en ella
    print('Total de valores nulos de', feature, '=', data_frame[feature].isna().sum())
```

### Observacion : Despues de analizar los datos nulos, podemos apreciar una gran cantidad faltante  en las columnas [Evaporation, Sunshine, Cloud9am, Cloud3pm], para no reducir nuestra muestra en base a la gran cantidad de datos nulos de estas columnas se procederan a omitir estas columnas en posteriores analisis.

```python
## Creamos un nuevo DataFrame llamado data_frame_2 que es igual al DataFrame original
## pero sin las columnas 'Evaporation', 'Sunshine', 'Cloud9am' y 'Cloud3pm'
data_frame_2 = data_frame.drop(['Evaporation', 'Sunshine', 'Cloud9am','Cloud3pm'], axis=1)

## Este bucle itera a través de todas las columnas del nuevo DataFrame (data_frame_2)
for feature in data_frame_2.columns:
    ## Imprime el nombre de la columna y el número de valores nulos en ella
    print('Total de valores nulos de', feature, '=', data_frame_2[feature].isna().sum())
```

```python
## Se crea un nuevo DataFrame 'data_frame_limpio' eliminando todas las filas que contienen valores nulos en el DataFrame 'data_frame_2'
data_frame_limpio = data_frame_2.dropna()

## Este bucle itera a través de todas las columnas del DataFrame 'data_frame_limpio'
for feature in data_frame_limpio.columns:
    ## Imprime el nombre de la columna y el número de valores nulos en ella (debería ser cero para todas las columnas en este punto)
    print('Total de valores nulos de', feature, '=', data_frame_limpio[feature].isna().sum())
```

# Hallazgo 1:Promedio de Temperatura Maxima y Minima por Año

```python
data_frame_limpio['Date'] = pd.to_datetime(data_frame_limpio['Date'], format='%Y-%m-%d')
```

```python

promedio_temp_max = data_frame_limpio.groupby(data_frame_limpio["Date"].dt.year)["MaxTemp"].mean()
promedio_temp_min = data_frame_limpio.groupby(data_frame_limpio["Date"].dt.year)["MinTemp"].mean()

plt.plot(promedio_temp_max.index, promedio_temp_max.values, label='Temperatura máxima')
plt.plot(promedio_temp_min.index, promedio_temp_min.values, label='Temperatura mínima')

plt.xlabel('Año')
plt.ylabel('Temperatura (°C)')
plt.title('Promedio de temperatura máxima y mínima por año')
plt.xticks(promedio_temp_max.index[::1]) 
plt.legend()

for i in range(len(promedio_temp_max)):
    plt.text(promedio_temp_max.index[i], promedio_temp_max.values[i], round(promedio_temp_max.values[i], 1))
    plt.text(promedio_temp_min.index[i], promedio_temp_min.values[i], round(promedio_temp_min.values[i], 1))
plt.show()
```

En este hallazgo podemos encontrar que a lo largo de los años la temperatura minima se ha mantenido relativamente estable a lo largo del 2007 al 2016. En cambio las temperaturas maximas tienen una variacion mas significativa. En ambos casos vemos que en el 2017 hubo un significativo cambio climatico en australia, este año ha sido uno de los 3 mas calurosos en la historia de australia segun los estudios.
